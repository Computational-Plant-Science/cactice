First, read in some sample grids from a CSV file and take a look at them.

```python
from cactice.io import read_grids_csv
from cactice.vis import show_grid

grids = read_grids_csv('../testdata/grids_sample1.csv')

for name, grid in grids.items():
    show_grid(grid, f"Grid {name}")
```

Next we can calculate the class distribution on a per-plot basis and across all plots.

```python
from pprint import pprint
import cactice.stats as stats

classes = stats.classes(grids.values())
print("Classes:")
pprint(classes)

transitions_horiz, transitions_vert = stats.undirected_transitions(grids.values())
print("Undirected transitions (horizontal):")
pprint(transitions_horiz)
print("Undirected transitions (vertical):")
pprint(transitions_vert)
```
